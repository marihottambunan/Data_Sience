```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
# Load your dataset from 'data.csv'
df = pd.read_csv('data.csv')

# Display the original dataset
print("Original Dataset:") 
print(df.head())
```

```python
if 'Programming Skill' in df.columns:
    # Perform Z-score Standardization
    scaler = StandardScaler() 
    df_standardized = pd.DataFrame(scaler.fit_transform(df[['Programming Skill']]), columns=['Programming Skill'])
    # Display the standardized dataset
    print("Standardized Dataset (Z-score Standardization)") 
    print(df_standardized.head())

    # Perform Min-Max Scaling
    scaler = MinMaxScaler() 
    df_scaled = pd.DataFrame(scaler.fit_transform(df[['Programming Skill']]), columns=['Programming Skill'])
    # Display the scaled dataset
    print("Scaled Dataset (Min-Max Scaling)") 
    print(df_scaled.head())

    # Perform Robust Scaling
    scaler = RobustScaler() 
    df_robust = pd.DataFrame(scaler.fit_transform(df[['Programming Skill']]), columns=['Programming Skill'])
    # Display the robustly scaled dataset
    print("Robustly Scaled Dataset (Robust Scaling)") 
    print(df_robust.head())
else:
    print("Kolom 'Programming Skill' tidak ditemukan dalam dataset.")
```
